```python
Eliminar Valores Atípicos (Outliers) con IQR
```

```python
import numpy as np

# Detectar valores atípicos con IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Definir límites de valores atípicos
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar datos dentro de los límites
df = df[~((df < limite_inferior) | (df > limite_superior)).any(axis=1)]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ver los nombres exactos de las columnas en el DataFrame
print("Columnas en el DataFrame:", df.columns.tolist())

# Nombres corregidos según lo que mencionaste
variables = [
    "FRAGILIDAD (FGL)", 
    "EFICIENCIA ADMINISTRATIVA (GA)", 
    "LIQUIDEZ (OP)", 
    "SOLVENCIA (PA)", 
    "RENTABILIDAD (ROE)", 
    "ACTIVOS NO RELACIONADOS AL GIRO DEL NEGOCIO (OA)", 
    "INVERSIONES (IC)", 
    "PROVISIONES DE CARTERA (PVC)",  
    "PIB", 
    "INFLACION (INF)", 
    "OFERTA MONETARIA (M1)"
]

# Limpiar nombres de columnas (eliminar espacios extra si existen)
df.columns = df.columns.str.strip()

# Filtrar solo las variables que están en el DataFrame
variables_presentes = [col for col in variables if col in df.columns]

if not variables_presentes:
    print("Ninguna de las variables está en el DataFrame. Verifica los nombres con print(df.columns)")
else:
    # Crear variables rezagadas (lag) para análisis de series temporales
    lags = 1  # Definir el número de períodos de rezago
    df_lagged = df.copy()
    
    for var in variables_presentes:
        df_lagged[f"{var}_lag{lags}"] = df[var].shift(lags)
    
    # Eliminar valores nulos generados por el rezago
    df_lagged = df_lagged.dropna()
    
    # Mostrar las primeras filas de los datos con lag
    print(df_lagged.head())
    
    # Calcular estadísticas descriptivas
    descriptive_stats = df_lagged[variables_presentes].describe().T
    descriptive_stats["Mediana"] = df_lagged[variables_presentes].median()
    
    # Mostrar la tabla con los resultados de manera clara en formato estándar
    print(descriptive_stats)
    
    # Visualización de estadísticas descriptivas en una tabla más legible
    fig, ax = plt.subplots(figsize=(22, 14))  # Aumentamos aún más el tamaño de la figura
    ax.axis('tight')
    ax.axis('off')
    table = ax.table(cellText=descriptive_stats.round(6).values,
                      colLabels=descriptive_stats.columns,
                      rowLabels=descriptive_stats.index,
                      cellLoc='center', loc='center', bbox=[0, 0, 1, 1])
    table.auto_set_font_size(False)
    table.set_fontsize(14)  # Aumentamos el tamaño de la fuente aún más
    for key, cell in table.get_celld().items():
        cell.set_height(0.05)  # Ajustamos la altura de las filas para mayor claridad
    plt.title("Estadísticas Descriptivas con Lag", fontsize=20, fontweight='bold')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score, roc_curve, auc
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Cargar los datos en un DataFrame (asumiendo que ya están en df)
# df = pd.read_csv("archivo.csv")  # Si los datos provienen de un archivo CSV

# Eliminar columnas de fecha irrelevantes
df = df.drop(columns=["DIA", "MES", "AÑO", "TRIMESTRE"], errors='ignore')

# Definir el porcentaje de división para entrenamiento
train_size = 0.8
split_index = int(len(df) * train_size)

# División secuencial respetando la temporalidad
train = df.iloc[:split_index]
test = df.iloc[split_index:]

# Separar variables predictoras (X) y variable objetivo (y)
# Suponiendo que "FRAGILIDAD (FGL)" es la variable objetivo
X_train = train.drop(columns=["FRAGILIDAD (FGL)"])
y_train = train["FRAGILIDAD (FGL)"]

X_test = test.drop(columns=["FRAGILIDAD (FGL)"])
y_test = test["FRAGILIDAD (FGL)"]

# Verificar las dimensiones de los conjuntos
print("Tamaño del conjunto de entrenamiento:", X_train.shape, y_train.shape)
print("Tamaño del conjunto de prueba:", X_test.shape, y_test.shape)

# Entrenar un modelo Random Forest para generar la matriz de confusión
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)
y_prob_test = clf.predict_proba(X_test)[:, 1]  # Probabilidades para la curva ROC

# Reporte de clasificación - Entrenamiento
print("Reporte de Clasificación - Entrenamiento:")
print(classification_report(y_train, y_pred_train))

# Matriz de confusión - Entrenamiento
cm_train = confusion_matrix(y_train, y_pred_train)
disp_train = ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=["No Frágil", "Frágil"])
disp_train.plot(cmap='Blues')
plt.title("Matriz de Confusión - Entrenamiento")
plt.show()

# Reporte de clasificación - Prueba
print("Reporte de Clasificación - Prueba:")
print(classification_report(y_test, y_pred_test))

# Matriz de confusión - Prueba
cm_test = confusion_matrix(y_test, y_pred_test)
disp_test = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=["No Frágil", "Frágil"])
disp_test.plot(cmap='Blues')
plt.title("Matriz de Confusión - Prueba")
plt.show()

# Cálculo del % de precisión
accuracy = accuracy_score(y_test, y_pred_test)
print(f"Precisión del modelo: {accuracy:.2%}")

# Curva AUC-ROC
fpr, tpr, _ = roc_curve(y_test, y_prob_test)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("Tasa de Falsos Positivos")
plt.ylabel("Tasa de Verdaderos Positivos")
plt.title("Curva AUC-ROC")
plt.legend()
plt.show()

# Importancia de las variables
importances = clf.feature_importances_
features = X_train.columns

# Ordenar las variables por importancia
df_importance = pd.DataFrame({'Variable': features, 'Importancia': importances})
df_importance = df_importance.sort_values(by='Importancia', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=df_importance['Importancia'], y=df_importance['Variable'], hue=df_importance['Variable'], dodge=False, legend=False, palette='viridis')
plt.xlabel("Importancia")
plt.ylabel("Características")
plt.title("Importancia de las Variables en Random Forest")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score, roc_curve, auc
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Cargar los datos en un DataFrame (asumiendo que ya están en df)
# df = pd.read_csv("archivo.csv")  # Si los datos provienen de un archivo CSV

# Verificar si las columnas de fecha existen antes de eliminarlas
columns_to_drop = ["AÑO", "MES", "DIA", "TRIMESTRE"]
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns], errors='ignore')

# Definir el porcentaje de división para entrenamiento
train_size = 0.8
split_index = int(len(df) * train_size)

# División secuencial respetando la temporalidad
train = df.iloc[:split_index]
test = df.iloc[split_index:]

# Separar variables predictoras (X) y variable objetivo (y)
# Suponiendo que "FRAGILIDAD (FGL)" es la variable objetivo
X_train = train.drop(columns=["FRAGILIDAD (FGL)"])
y_train = train["FRAGILIDAD (FGL)"]

X_test = test.drop(columns=["FRAGILIDAD (FGL)"])
y_test = test["FRAGILIDAD (FGL)"]

# Crear categorías de rangos (Bajo, Medio, Alto) para cada variable independiente
df_bivar = df.copy()
for variable in X_train.columns:
    try:
        df_bivar[variable + "_Rango"] = pd.qcut(df_bivar[variable].rank(method='first'), q=3, labels=["Bajo", "Medio", "Alto"], duplicates="drop")
    except ValueError:
        print(f"No se pudo crear rangos para {variable} debido a valores repetidos o insuficientes.")

# Configurar el tamaño del gráfico
num_vars = len(X_train.columns)
filas = (num_vars // 3) + (num_vars % 3 > 0)
fig, axes = plt.subplots(nrows=filas, ncols=3, figsize=(15, 5 * filas))
fig.suptitle("Relación entre Variables Independientes y Fragilidad (FGL)", fontsize=14, fontweight='bold')

# Colores personalizados
colors = ["blue", "orange"]

# Graficar cada variable independiente contra la fragilidad financiera con porcentajes
for ax, variable in zip(axes.flat, X_train.columns):
    if variable + "_Rango" in df_bivar.columns:
        tabla = pd.crosstab(df_bivar[variable + "_Rango"], df_bivar['FRAGILIDAD (FGL)'], normalize='index')
        tabla.plot(kind='bar', stacked=True, color=colors, ax=ax, width=0.8, legend=False)
        ax.set_title(variable, fontsize=10)
        ax.set_xlabel("Rangos (Bajo, Medio, Alto)")
        ax.set_ylabel("Proporción (%)")
        
        # Agregar etiquetas de porcentaje dentro de las barras
        for container in ax.containers:
            for bar in container:
                height = bar.get_height()
                if height > 0:
                    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2,
                            f"{height:.1%}", ha='center', va='center', fontsize=8, color='white', fontweight='bold')
    else:
        ax.axis('off')

# Ajustar la distribución del gráfico
handles = [plt.Rectangle((0,0),1,1, color=color) for color in colors]
fig.legend(handles, ["FGL = 0 (No Frágil)", "FGL = 1 (Frágil)"], loc='upper center', ncol=2, fontsize=10)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
import os

# Verificar si el archivo existe en el directorio actual
file_path = "PRUEBA3CONFECHAS.csv"
if not os.path.exists(file_path):
    raise FileNotFoundError(f"El archivo {file_path} no se encuentra en el directorio actual.")

# Cargar los datos
df = pd.read_csv(file_path)

# Eliminar espacios en los nombres de las columnas
df.columns = df.columns.str.strip()

# Convertir FECHA a datetime y extraer características temporales
if 'FECHA' in df.columns:
    df['FECHA'] = pd.to_datetime(df['FECHA'], errors='coerce')
    df['AÑO'] = df['FECHA'].dt.year
    df['MES'] = df['FECHA'].dt.month
    df['DIA'] = df['FECHA'].dt.day
    df['TRIMESTRE'] = df['FECHA'].dt.quarter
    df = df.drop(columns=['FECHA'])  # Eliminar la columna FECHA

# Verificar y eliminar columnas no numéricas (excepto FRAGILIDAD)
non_numeric_columns = df.select_dtypes(exclude=['number']).columns
if not non_numeric_columns.empty:
    df = df.drop(columns=non_numeric_columns)

# Separar variables independientes (X) y dependiente (Y)
X = df.drop(columns=['FRAGILIDAD (FGL)'])
Y = df['FRAGILIDAD (FGL)']

# Convertir todas las variables numéricas a tipo float
X = X.astype(float)

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)

# Entrenar el modelo Random Forest
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train, Y_train)

# Configuración del tamaño de la figura
plt.figure(figsize=(20, 10))

# Visualización de uno de los árboles dentro del modelo Random Forest
plot_tree(modelo_rf.estimators_[0], feature_names=X.columns, class_names=["No Frágil", "Frágil"], filled=True, fontsize=8)

# Título del gráfico
plt.title("Visualización del Árbol de Decisión en Random Forest")

# Mostrar el gráfico
plt.show()
```
